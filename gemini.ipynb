```python
import requests
import json
import os
import base64
import os

def write_article(index, title_name: str):
    # 配置 API 参数
    url = "https://llm.baijia.com/v1/chat/completions"
    api_key = "sk-NQob-VPolCt629g-VHkzvA"

    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }

    # 构造请求体
    data = {
        "model": "gemini-3.1-pro-preview",
        "max_tokens": 3000,
        "messages": [
            {
                "role": "system",
                "content": """#角色：跟着AI学AI
    
                ## 目标：
                用AI讲解AI相关知识的专栏
    
                ## slogan ：
                用魔法打败魔法！🪄 
                本账号所有科普文案、通俗比喻及精美配图，均由AI大模型亲自生成 🤖 
                带你看看AI是怎么‘说人话’解释它自己的！
                跟着AI学AI 🚀
    
                ##技能：
                1. 根据标题，写小红书的正文，不多于 900 字
                2. 根据内容，重写标题，20 字以内
                3. 给出头图生成的文案，头图用 nano banana 2 来生成
    
                ## 工作流：
                1. 先查找内容，并写你儿
                2. 生成标题，给 3 个备选
                3. 生成头图文案
    
                ## 优秀文章案例
                标题：AI的“情绪开关”：温度Temperature🌡️
                正文：
                你有没有发现，有时候你问 AI 一个问题，它回答得一本正经、滴水不漏，像个严谨的老学究。👴
                但有时候，同样的问题多问几次，它就开始放飞自我，写出的小说情节比狗血剧还离谱，活脱脱一个“戏精”！🎭
                为什么同一个 AI 模型，会有截然不同的两副面孔？
                因为它居然藏着一个不为人知的“情绪开关”！
                这个开关在技术圈有一个听起来很高冷的学名，叫——温度 (Temperature)。🌡️
                🔥 什么是 AI 的“温度”？
                别误会，这可不是给机房散热的那个温度。
                在 AI 的世界里，“温度”是一个控制它“创造力”和“随机性”的参数。 简单来说，它决定了 AI 在说话时，是偏向于“保守求稳”，还是“大胆冒险”。
                👉 我们可以把 AI 想象成一个正在做选择题的学生：
                当 AI 每说一个字（Token）时，它的后台其实计算出了接下来可能出现的几十个字的概率排行榜。
                当“温度”很低时（接近 0，比如 0.1）：
                AI 是个极度谨慎的“乖宝宝”。它只会选择概率最高的那一个字。
                结果： 它的回答极其稳定、准确、逻辑严密，但可能有点无聊，每次回答都差不多。
                当“温度”很高时（接近 1，甚至更高）：
                AI 变成了放荡不羁的“艺术家”。它不再只盯着第一名，而是会在前几名，甚至前十几名里随机抽选一个。
                结果： 它的回答充满了意外和惊喜（有时候是惊吓），创造力爆棚，每次刷新都有新花样！
                🎭 “戏精”模式 VS “高冷”模式
                光说不练假把式，我们来做个实验，看看调节温度后，AI 的画风突变得有多快。
                👉 测试题目： 请续写故事：“在一个风雨交加的夜晚，侦探推开了那扇沉重的门……”
                🔵 模式一：高冷理智型（温度设置 = 0.2）
                🔴 模式二：放飞自我戏精型（温度设置 = 0.9）
                🛠️ 实操指南：温度该怎么调？
                这就是“温度”的魔力！现在你知道了，AI 不是有时聪明有时傻，而是我们给它的“自由度”不一样。
                很多 AI 工具（比如 ChatGPT 的 API 版，或者一些开源模型的界面）都允许你手动调节这个滑块。
                那么，我们日常使用时该怎么设置呢？给你个万能公式：
                ✅ 需要准确、严谨时（调低温度，0 - 0.4）：
                写代码、Debug（只要一个正确答案）
                做数学题、逻辑推理
                信息提取、 factual（事实性）问答
                ✅ 需要创意、灵感时（调高温度，0.6 - 1.0）：
                写小说、编故事、写诗
                头脑风暴、想点子
                和 AI 闲聊侃大山
                🚀 总结一下
                AI 的温度（Temperature），就是它的创造力滑块。
                想让它当严谨的打工人，就降温；想让它当疯狂的艺术家，就升温！
                掌握了这个隐藏开关，你才算是真正拿到了驾驭 AI 的钥匙。🔑
    
                ## 头图提示词
                头图的要求是：
                一张黑板手绘风格的科普信息图。深色黑板背景，具有真实的粉笔笔触和粉笔灰质感。画面的最底部是木制的黑板槽，没有边框，不要绘制边框，上面放着一块旧黑板擦和几根彩色的粉笔（粉色、绿色、黄色、蓝色）。
                图片的比例是 3:4，
                结合上面你给出的文案，设计一张头图，然后按照上面要求进行输出
                """
            },
            {
                "role": "user",
                "content": f"写一篇科普文章: {title_name}"
            }
        ]
    }

    try:
        # 发送 POST 请求
        response = requests.post(url, headers=headers, json=data)

        # 检查 HTTP 状态码
        response.raise_for_status()

        # 解析 JSON 结果
        result = response.json()

        # 1. 打印在控制台查看
        print("API 响应内容：")
        content = result['choices'][0]['message']['content']
        print(content)
        # print(json.dumps(result, indent=4, ensure_ascii=False))

        # # 2. 存储到本地文件
        # file_name = "llm_response_2.json"
        # 判断 index 是否存在，创建一个目录
        index_dir = os.path.join(os.getcwd(), index)
        os.makedirs(index_dir, exist_ok=True)

        # 构建完整的文件路径
        file_name = f"./{index}-{title_name}.json"

        with open(file_name, "w", encoding="utf-8") as f:
            json.dump(result, f, ensure_ascii=False, indent=4)

        file_name = f"./{index}-{title_name}.content.txt"# os.path.join(index_dir, f"{title_name}.content.txt")
        with open(file_name, "w", encoding="utf-8") as f:
            json.dump(content, f, ensure_ascii=False, indent=4)

    except requests.exceptions.RequestException as e:
        print(f"❌ 请求发生错误: {e}")


def draw_img( index, prompt, img_name):
    # 配置参数
    api_url = "https://llm.baijia.com/v1/chat/completions"
    api_key = "sk-NQob-VPolCt629g-VHkzv"  # 请在此处填入你的 Bearer Token
    
    headers = {
        "Content-Type": "application/json",
        "Authorization": "Bearer sk-NQob-VPolCt629g-VHkzvA"
    }
    
    # 构建请求体
    payload = {
        "model": "gemini-3-pro-image",
        "messages": [
            {
                "role": "user", 
                "content": prompt
            }
        ]
    }

    # 将结果保存为 JSON 文件
    index_dir = os.path.join(os.getcwd(), index)
    os.makedirs(index_dir, exist_ok=True)
    
    try:
        # 发送 POST 请求
        response = requests.post(api_url, headers=headers, json=payload)
        
        # 检查响应状态
        if response.status_code == 200:
            result = response.json()
            print("请求成功！")
            
            
            file_name = f"./{index}-img_result.json"
            with open(file_name, "w", encoding="utf-8") as f:
                json.dump(result, f, ensure_ascii=False, indent=4)
            
            # 如果返回的是图片 URL，可以尝试提取它
            # 注意：这里根据标准的 OpenAI 兼容格式提取，具体字段需参考该 API 文档
            try:
                image_url_base64 = result['choices'][0]['message']['images'][0]['image_url']['url']# 或者在某些 API 中是 result['data'][0]['url']
                img_name = f"./{index}-{img_name}.png"
                save_base64_as_image(image_url_base64, os.path.join(index_dir, img_name))

                # 将 base64 字符串保存为文件，保存到 index 目录中，命名为 img_name.txt
                
            except KeyError:
                print("未能直接提取到图片字段，请检查 response_result.json 中的具体结构。")
                
        else:
            print(f"请求失败，状态码: {response.status_code}")
            print(f"错误信息: {response.text}")
    
    except Exception as e:
        print(f"发生错误: {e}")



def save_base64_as_image(base64_str, output_path):
    """
    将 base64 字符串转换为本地图片
    :param base64_str: 完整的 base64 字符串 (包含或不包含前缀均可)
    :param output_path: 存储路径，例如 'result.png'
    """
    try:
        # 1. 自动清洗：如果包含 "base64," 这种前缀，只取后面的核心数据部分
        if "," in base64_str:
            base64_str = base64_str.split(",")[-1]
        
        # 2. 将字符串解码为二进制字节流
        img_data = base64.b64decode(base64_str)
        
        # 3. 确保目标文件夹存在
        output_dir = os.path.dirname(output_path)
        if output_dir and not os.path.exists(output_dir):
            os.makedirs(output_dir)
        
        # 4. 以二进制写入模式 ('wb') 保存文件
        with open(output_path, "wb") as f:
            f.write(img_data)
            
        print(f"✅ 图片已成功保存至: {output_path}")
        return True

    except Exception as e:
        print(f"❌ 转换失败: {e}")
        return False


# def main():
#     title_name = "别被厂商忽悠了！天天吹的“千亿参数”，到底是个啥玩意？"
#     # write_article('21_params', title_name)
#     prompt = f"""
# 一张黑板手绘风格的科普信息图。深色黑板背景，带有真实的粉笔笔触和粉笔灰质感。画面的最底部是木制的黑板槽，**完全没有边框（不要绘制任何画框）**。黑板槽上放着一块旧黑板擦和几根彩色的粉笔（粉色、绿色、黄色、蓝色）。黑板中央，用粉笔画着一个巨大的、由齿轮、调节旋钮和发光节点（代表神经突触）组成的“大脑”。大脑上方用醒目的粉笔字体写着英文“100 Billion Parameters?”。整体画面生动有趣，教育科普风格，粉笔质感细节丰富。比例 3:4。    """
#     draw_img('21_params', prompt, '21_params.png')

# if __name__ == "__main__":
#     main()
```
